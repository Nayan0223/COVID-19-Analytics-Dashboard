```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('COVID-19 Global Statistics Dataset.csv')
```

```python
df.head(6)
```

```python
df.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer

# Create a copy of the DataFrame to avoid modifying the original
df_imputed = df.copy()

# Initialize the imputer for numerical columns
num_imputer = SimpleImputer(strategy='mean')

# Select numerical columns for imputation
num_cols = df_imputed.select_dtypes(include=[np.number]).columns

# Fit and transform the numerical columns
df_imputed[num_cols] = num_imputer.fit_transform(df_imputed[num_cols])

# For categorical columns, we can use the most frequent value
cat_imputer = SimpleImputer(strategy='most_frequent')

# Select categorical columns for imputation
cat_cols = df_imputed.select_dtypes(include=[object]).columns

# Fit and transform the categorical columns
df_imputed[cat_cols] = cat_imputer.fit_transform(df_imputed[cat_cols])
```

```python
print("Missing values after filling:")
print(df_imputed.isnull().sum())
```

```python
print("Cleaned DataFrame:")
print(df_imputed)
```

```python
df.to_csv('clean-covid1.csv',index=False)
```
